```python
import pandas as pd

ds = pd.read_csv(r'C:\Users\DELL\Desktop\student_scores.csv')
df = ds.copy() 
x = df.iloc[:,:-1]  # Select all columns except the last one
y = df.iloc[:,-1]  # Select the last column as the target variable

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)

from sklearn.linear_model import LinearRegression
# Create a linear regression model
model = LinearRegression()

# Fit the model to the training data
model.fit (x_train, y_train)

# Making predictions on the test data
y_pred = model.predict(x_test)

from sklearn.metrics import r2_score
# Calculate the R-squared score
r2 = r2_score(y_test, y_pred)
print(f"R-squared score: {r2:.2f}") 

# Print the coefficients of the linear regression model (slope(b1) and intercept(b0))
model.slope = model.coef_
model.bias = model.intercept_

# The equation of the linear regression model is: y = 2.79 + 9.71*x
print(f"Slope (b1): {model.slope}")
print(f"Intercept (b0): {model.bias}")

import math
from sklearn.metrics import mean_squared_error
# Calculate the root mean squared error
rmse = math.sqrt(mean_squared_error(y_test, y_pred))
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")

# Plotting the results
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")  
plt.figure(figsize=(5, 3)) 
plt.scatter(x_test, y_test, label='Actual')  
plt.plot(x_test, y_pred, color='red', label='Predicted')  
plt.ylabel('Scores')
plt.title('Actual vs Predicted Scores')
plt.ylim(ymin=0)
plt.tight_layout()
plt.legend() # show the labels of the plot
plt.show()

```
